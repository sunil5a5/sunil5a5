```python
x= data.drop(["User_ID", "Calories"], axis= 1)
```

```python
y= data.Calories
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, LabelEncoder
encoder= LabelEncoder()
```

```python
x["Gender"]= encoder.fit_transform(x[["Gender"]])
```

```python
x.head()
```

```python
x.Gender.value_counts()
```

```python
Scaler= MinMaxScaler()
```

```python
for a in x.columns[1:]:
  x[a]= Scaler.fit_transform(x[[a]])
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42)
```

```python
x_train.head()
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_regression
```

```python
kbest= SelectKBest(mutual_info_regression, k=4)
```

```python
kbest.fit(x_train, y_train)
```

```python
x_train_new= x_train[x_train.columns[kbest.get_support()]]
```

```python
x_test_new= x_test[x_test.columns[kbest.get_support()]]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
from xgboost import XGBRegressor
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
def model_performance(model):
  model.fit(x_train_new, y_train)
  ans= model.predict(x_test_new)
  error= mean_absolute_error(y_test, ans)
  return error
```

```python
model_performance(LinearRegression())
```

```python
model_performance(RandomForestRegressor())
```

```python
model_performance(AdaBoostRegressor())
```

```python
model_performance(XGBRegressor())
```

```python
model_performance(RandomForestRegressor(n_estimators=1000))
```

```python
model_performance(KNeighborsRegressor())
```

```python
model = RandomForestRegressor()
model.fit(x_train_new, y_train)
ans= model.predict(x_test_new)
error= mean_absolute_error(y_test, ans)
```
